# Básicos do matplotlib
----

O matplotlib é a biblioteca de plotagem mais robusta e "influencial" do python, ou seja, muitos principios dela são adotados por outras bibliotecas. 

Ela possui vários módulos e funcionalidades. No entanto, em geral, interagimos com ela via o módulo pyplot.

```python
# Demo: Importando pyplot

import matplotlib.pyplot as plt
```

## Plot de linha (dois estilos)

O matplotlib pode ser usado de duas formas:

- Interface semelhante ao matlab
- Interface orientada a objetos

```python
# Demo: Plot de linha usando o estilo de interface do matlab

x = [i for i in range(-10, 11)]
y1 = [i**3 for i in x]
y2 = [i**4 for i in x]

plt.plot(x, y1)
```

> Dica: No matplotlib podemos usar estilos diferentes!

```python
# Demo: Usando um estilo diferente!

plt.style.use("seaborn")
```

```python
# Demo: Plot de linha usando interface orientada a objetos (criando figura separadamente)

fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y1)
```

```python
# Demo: Plot de linha usando interface orientada a objetos (usando subplots)

fig, ax = plt.subplots()
ax.plot(x, y1)
```

> Dica: O modo de plotagem do matlab é ótimo para criar visualizações rápidas. No entanto o estilo orientado a objeto é mais poderoso e natural de usar dentro do Python.

## Backends

No jupyter podemos usar as "magics". Comandos que simplificam a forma que interagimos com o python. E existe uma magic que nos ajuda a mudar o backend do matplotlib.

Nos cadernos jupyter, estaremos interessados em duas backend:

- inline
- notebook

O padrão atual é o inline. Ela nos dará as figuras de modo estático dentro do notebook.

Se quisermos as figuras no modo interativo. Precisamos usar o backend "notebook".

## Detalhes do plot

Podemos ajustar todos os elementos de um plot no matplotlib. Vamos ver alguns detalhes básicos:

- Título
- Etiquetas
- Marcadores
- Estilos de linha
- Cores

```python
# Demo: Customizando plot

plt.plot(x, y1, ls="-", marker="<", c="forestgreen")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Curva $x^3$")
```

> Dica: Você pode usar fórmulas matematicas do LaTex dentro do matplotlib

```python
help(plt.plot)
```

## Sobreposição de plots

Plots podem ser sobrepostos num mesmo eixo! Podemos usar uma legenda pra diferenciar os elementos.

```python
# Demo: Sobreposição de plots e legenda

fig, ax = plt.subplots()

# Sobrepondo plots num único eixo
ax.plot(x, y1, label="cubos")
ax.plot(x, y2, label="à quarta")
ax.legend()
```

## Plots compostos

Podemos incluir vários plots numa unica figura!

```python
# Demo: Plots em linha e plots em coluna

fig, [ax1, ax2] = plt.subplots(nrows=2, sharex=True)

# Eixo 1
ax1.plot(x, y1)
ax1.set_title("Cubos")
ax1.set_ylabel("Eixo Y")

# Eixo 2
ax2.plot(x, y2)
ax2.set_title("a quarta")
ax2.set_xlabel("Eixo X")
ax2.set_ylabel("Eixo Y")
```

> Dica: Note que na célula acima os valores retornados como eixo foram desempacotados.

## Scatter plots

Usamos até então o comando "plot" que também pode ser usado para fazer scatter plots. No entando usando o comando "scatter" temos muito mais controle do parâmetros!

```python
# Demo: Scatter plot com "plot"

plt.plot(x, y1, marker="o", ls="None")
```

```python
# Demo scatter com tamanho e cor ajustados
import matplotlib.colors as colors

color = [abs(i) for i in y1]
size = [abs(i) for i in y1]

plt.scatter(x, y1, c=color, s=size, norm=colors.LogNorm(), cmap="viridis")
plt.colorbar()
```

## Imagens

O matplotlib também é muito útil para visualizar imagens e mapas.

```python
# O estilo default é melhor para ver imagens ...
plt.style.use("default")
```

```python
# Demo: abrindo e mostrando uma imagem astronômica

# Carregando imagem astronomica no formato FITS
from astropy.io import fits
imagem = fits.getdata("OI2019A-031_2019-07-08_0234.fits")

# Mostrando imagem com imshow
plt.imshow(imagem, cmap="gray", norm=colors.LogNorm())
plt.colorbar()
```
